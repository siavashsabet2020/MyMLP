```python
import preprocess as ppc
import MLPClassifier
from MLPClassifier import List_AccTrain, List_MseTrain, List_AccValid, List_MseValid
import pandas as pd
import savemlp
```

```python
# Pre Process Object
file = pd.read_csv('../Datasets/cancer_.csv')
# file = pd.read_csv('boston.csv')
# preproc_data = ppc.Preprocess(file, 13, 13)
preproc_data = ppc.Preprocess(file, 9, 9)
print(preproc_data.inputs)
```

```python
shuffled_data = preproc_data.shuffle()
normalized_data = preproc_data.normalize('minmax')
preproc_data.split()
```

```python
x_train, y_train = preproc_data.X_train, preproc_data.Y_train
x_test, y_test = preproc_data.X_val, preproc_data.Y_val
```

```python
MyMlp = MLPClassifier.MLP([9, 2, 4, 1], activation='sigmoid', learning_rate_init=0.01,learn_rate='adaptive', verbose=True, tolerance=10**-1000, momentum=0.009)

out = MLPClassifier.train(MyMlp, x_train, y_train, x_test, y_test)
out = List_AccTrain, List_MseTrain, List_AccValid, List_MseValid
```

```python
savemlp.save_mlp(MyMlp, 'KiarashMLP', x_train, y_train)
```
